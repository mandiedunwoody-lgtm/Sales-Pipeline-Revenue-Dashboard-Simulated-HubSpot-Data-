```python
import pandas as pd

# Load raw deals file
deals = pd.read_excel(r"C:\Users\cutes\OneDrive\Desktop\deals.xlsx")

# Standardize column names
deals.columns = deals.columns.str.lower().str.strip()

# Convert dates
if "close_date" in deals.columns:
    deals["close_date"] = pd.to_datetime(deals["close_date"], errors="coerce")

# Clean numeric fields
numeric_cols = ["amount", "probability"]
for col in numeric_cols:
    if col in deals.columns:
        deals[col] = pd.to_numeric(deals[col], errors="coerce")

# Strip whitespace from text fields
text_cols = ["deal_name", "owner", "stage"]
for col in text_cols:
    if col in deals.columns:
        deals[col] = deals[col].astype(str).str.strip()

# Save cleaned file
deals.to_csv("../data_clean/deals_clean.csv", index=False)

deals.head()


```

```python
import pandas as pd

deals = pd.read_excel(r"C:\Users\cutes\OneDrive\Desktop\deals.xlsx")
deals.head()

```

```python
deals = deals.drop_duplicates()
deals.shape
```

```python
deals["deal_stage"] = (
    deals["deal_stage"]
    .str.strip()
    .str.lower()
    .replace({
        "qualified to buy": "qualified",
        "presentation scheduled ": "presentation scheduled",
        "closed won ": "closed won",
        "closed lost ": "closed lost",
    })
)

deals["deal_stage"].value_counts()
```

```python
# Remove currency symbols and commas
deals["amount"] = (
    deals["amount"]
    .astype(str)
    .str.replace(r"[^0-9.]", "", regex=True)
    .astype(float)
)

deals["amount"].describe()
```

```python
deals["close_date"] = pd.to_datetime(deals["close_date"], errors="coerce")

# Remove rows with invalid dates
deals = deals.dropna(subset=["close_date"])
deals.shape
```

```python
import os
os.makedirs("data_clean", exist_ok=True)

```

```python
deals.to_csv("data_clean/deals_clean.csv", index=False)
```
