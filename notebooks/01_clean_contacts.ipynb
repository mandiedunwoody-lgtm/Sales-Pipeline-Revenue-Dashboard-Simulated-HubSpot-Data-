```python
# Clean column names: lowercase, replace spaces with underscores
contacts.columns = (
    contacts.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
)

contacts.head()
```

```python
contacts = contacts.drop_duplicates()
contacts.shape
```

```python
# Strip whitespace, lowercase email addresses
contacts["email"] = contacts["email"].str.strip().str.lower()

# Remove rows with missing or invalid emails
contacts = contacts[contacts["email"].str.contains("@", na=False)]
contacts.shape

```

```python
import pandas as pd

# Load raw contacts file
contacts = pd.read_excel(r"C:\Users\cutes\OneDrive\Desktop\contacts.xlsx")

# Standardize column names
contacts.columns = contacts.columns.str.lower().str.strip()

# Convert dates
contacts["created_date"] = pd.to_datetime(contacts["created_date"], errors="coerce")

# Strip whitespace from text fields
text_cols = ["first_name", "last_name", "email", "company", "job_title", "lifecycle_stage", "owner"]
for col in text_cols:
    contacts[col] = contacts[col].astype(str).str.strip()

# Save cleaned file
contacts.to_csv("../data_clean/contacts_clean.csv", index=False)

contacts.head()
```

```python
# Normalize lifecycle stage values
contacts["lifecycle_stage"] = (
    contacts["lifecycle_stage"]
    .str.strip()
    .str.lower()
    .replace({
        "lead ": "lead",
        "Lead": "lead",
        "qualified lead": "qualified",
        "customer ": "customer",
    })
)

contacts["lifecycle_stage"].value_counts()
```
